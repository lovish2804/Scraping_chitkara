## **FOR ALL THE PAGES**

```python
final=pd.DataFrame()
for j in range(1,9):
  webpage=requests.get('https://www.chitkara.edu.in/can/alumni-testimonials/page/{}/'.format(j)).text
  soup=BeautifulSoup(webpage,'html.parser')
  a=soup.find_all('div',class_='profileCon')
  name=[]
  degree=[]
  y=[]

  for i in a:
    try:
       name.append(i.h3.text)
    except:
       name.append(np.nan)
    try:
      d=i.p.p.span
      if(type(d)==type(None)):
        degree.append(np.nan)
      else:
        degree.append(d.text)
    except:
      degree.append(np.nan)
    try:
      y.append(i.p.p.text)
    except:
      y.append(np.nan)
  for i in range(len(y)):
    y[i]=str(y[i])
    y[i]=y[i].split(str(degree[i]))
  z=[]
  for i in y:
    try:
      z.append(i[1])
    except:
      z.append(np.nan)
  q=[]
  for i in z:
    i=str(i)
    q.append(i.split("\n"))
  test=[]
  try:
    for foo in soup.find_all('div', attrs={'class': 'modal-body'}):
      foo_descendants = foo.descendants
      for d in foo_descendants:
          if d.name == 'div' and d.get('class', '') == ['col-md-12']:
              test.append(d.p.text)
  except:
    test.append(np.nan)
    
  dff=pd.DataFrame(q)
  dff.insert(0,"Name",name)
  dff.insert(1,"Degree",degree)
  dff.rename(columns = {1 : 'Batch', 2 : 'Designation',4:'del'}, inplace = True)
  dff.drop(dff.columns[2], axis=1, inplace=True)
  dff.drop(dff.columns[4], axis=1, inplace=True)
  dff.insert(4,"Testimonial",test)
 

  final=final.append(dff,ignore_index=True)
final.drop('del', axis=1, inplace=True)
```
